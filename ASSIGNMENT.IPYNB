```python
#3. Write a Python program to count the occurrences of each word in a givensentence
sentence = input("Enter a sentence: ")
word_counts = {}
for word in sentence.split():
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1
print(word_counts)
```

```python
#6. Write a Python program to find the first appearance of the substring 'not' and'poor' 
# from a given string, if 'not' follows the 'poor', 
# replace the whole 'not'...'poor' substring with 'good'. Return the resulting string
string = input("Enter a string: ")
notstr = string.find("not")
poorstr = string.find("poor")  
if notstr != -1 and poorstr != -1 and notstr < poorstr:
    string = string[:notstr] + "good" + string[poorstr + 4:]
print(string)

```

```python
#7. Program to find Greatest Common Divisor of two numbers. For example, theGCD of 20 and 28 is 4 and the GCD of 98 and 56 is 14.
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

while num2 != 0:
    num1, num2 = num2, num1 % num2

print("GCD =", num1)    
```

```python
#8. Write a Python program to check whether a list contains a sublist.
def sublist(main_list, sub_list):
    for i in range(len(main_list) - len(sub_list) + 1):
        if main_list[i:i + len(sub_list)] == sub_list:
            return True
    return False


sublist([1, 2, 3],[3,4,5])
```

```python
#13.Write a Python program to sort a dictionary (ascending /descending) by value
def sort_dict(d, ascending=True):
    return dict(sorted(d.items(), key=lambda item: item[1], reverse=not ascending))
```

```python
#15.Given a number n, write a python program to make and print the list of Fibonacci
#series up to n. Input : n=7 Hint : first 7 numbers in the series Expected output :
#First few Fibonacci numbers are 0, 1, 1, 2, 3, 5, 8, 13
def fibonacci(n):
    fib_series = []
    a, b = 0, 1
    for _ in range(n):
        fib_series.append(a)
        a, b = b, a + b
    return fib_series

fibonacci(7)
```

```python
#16.Counting the frequencies in a list using a dictionary in Python. Input : [1, 1, 1, 5, 5, 3, 1, 3, 3, 1,4, 4, 4, 2, 2, 2, 2]
#Expected output : 1 : 5 , 2 : 4 , 3 : 3 , 4 : 3 , 5 : 2

def count_frequencies(lst):
    frequency_dict = {}
    for item in lst:
        if item in frequency_dict:
            frequency_dict[item] += 1
        else:
            frequency_dict[item] = 1
    return frequency_dict

count_frequencies([1,1,1,5,5,3,1,3,3,1,4,4,4,2,2,2,2])
```

```python
#17.Write a python program using function to find the sum of odd series andevenseries
#Odd series: 12/ 1! +32/ 3! + 52/ 5!+……n
#Even series: 22/ 2! + 42/ 4! + 62/ 6!+……n
import math

def odd_series(n):
    sum_odd = 0
    for i in range(1, n+1, 2):
        sum_odd += (i**2) / math.factorial(i)
    return sum_odd

def even_series(n):
    sum_even = 0
    for i in range(2, n+1, 2):
        sum_even += (i**2) / math.factorial(i)
    return sum_even

n = int(input("Enter value of n: "))

odd_sum = odd_series(n)
even_sum = even_series(n)

print("Sum of Odd Series =", odd_sum)
print("Sum of Even Series =", even_sum)
```

```python
#18. Python Program to Find Factorial of Number Using Recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
factorial(5)
```
