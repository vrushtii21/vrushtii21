```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('house_price_regression_dataset.csv')
df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

scaled_df.head()
```

```python
scaled_df.describe()
```

```python
x = scaled_df.drop(columns=['House_Price'])
y = scaled_df['House_Price']
print(x)
```

```python
x.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
Ypred = model.predict(x_test)
```

```python
New = x_test
New['House_Price'] = y_test
New['pred_House_Price'] = Ypred
```

```python
New
```

```python
New.to_csv("House_Price_ans.csv")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, root_mean_squared_error
```

```python
print(mean_squared_error(y_test,Ypred))
```

```python
print(mean_absolute_error(y_test,Ypred))
```

```python
print(root_mean_squared_error(y_test,Ypred))
```

```python
print(mean_absolute_percentage_error(y_test,Ypred))
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,Ypred))
```
