```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Engine Fuel Type','Market Category','Vehicle Size','Number of Doors'], axis=1)
```

```python
df.isnull().sum()
```

```python
duplicaterows = df[df.duplicated()]
print("number of duplicate rows: ", duplicaterows.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df
```

```python
df.count()
```

```python
df = df.dropna()    
df.count()
```

```python
df.isnull().sum()
```

```python
plt.bar(df['Year'], df['MSRP'])
plt.figsize=(10,10)
plt.xlabel('Year')
plt.ylabel('MSRP')
plt.title('MSRP vs Years')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
unique_brands = df['Make'].unique()
print(unique_brands)
```

```python
unique_brands.size
```

```python
brand_counts = df['Make'].value_counts()

plt.figure(figsize=(10,12))
plt.barh(brand_counts.index, brand_counts.values)
plt.xlabel("Number of Cars")
plt.ylabel("Brand")
plt.title("Cars per Brand")
plt.show()
```
